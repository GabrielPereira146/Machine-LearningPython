```python
nulo = data.isnull()
tem_nulo = nulo.any(axis=1)
tem_nulo
data_nulo = data[tem_nulo]
data_nulo
```

```python
data.dropna(thresh=len(data.columns) - 1, inplace=True)
data.info()
```

```python
##cat = data.select_dtypes(include = ['object']).columns
data['A1']=data['A1'].fillna(data['A1'].mode())
data.info()
```

```python
from sklearn.impute import SimpleImputer
cat = data.select_dtypes(include = ['object']).columns
imp = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
dados_2 = imp.fit_transform(data[cat])
dados_2
```

```python
cat = data.select_dtypes(include = ['float64','int64']).columns
imp = SimpleImputer(missing_values=np.nan, strategy='mean')
dados_3 = imp.fit_transform(data[cat])
dados_3
```

```python
dados = np.concatenate((dados_3,dados_2), axis=1)
dados
```

```python
data=pd.DataFrame(dados,columns=['A2','A3','A8','A11','A14','A15','A1','A4','A5','A6','A7','A9','A10','A12','A13','CLASSE'])
data['A2'] = data['A2'].apply(float)
data['A3'] = data['A3'].apply(float)
data['A8'] = data['A8'].apply(float)
data['A11'] = data['A11'].apply(int)
data['A14'] = data['A14'].apply(float)
data['A15'] = data['A15'].apply(int)
data.info()
```

```python
cat = data.select_dtypes(include = ['object']).columns
cat_ex = []
for col in cat:
    if(data[col].value_counts() >= 3):
        cat_ex.append(col)
cat_ex
```

```python
import pandas as pd
import numpy as np

data = pd.read_csv('crx.csv', na_values='?')
data.head()#printa os 5 primeiros registros
```

```python
cat = data.select_dtypes(include = ['object']).columns

for col in cat:
    print(data[col].value_counts())
```

```python
cat2 = data.select_dtypes(include = ['float64','int64']).columns

for col in cat2:
    print(col,":",data[col].mean())
```

```python
nulo = data.isnull()
tem_nulo = nulo.any(axis=1)
tem_nulo
data_nulo = data[tem_nulo]
data_nulo
```

```python
data.dropna(thresh=len(data.columns) - 1, inplace=True)
data.info()
```

```python
##cat = data.select_dtypes(include = ['object']).columns
data['A1']=data['A1'].fillna(data['A1'].mode())
data.info()
```

```python
from sklearn.impute import SimpleImputer
cat = data.select_dtypes(include = ['object']).columns
imp = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
dados_2 = imp.fit_transform(data[cat])
dados_2
```

```python
cat = data.select_dtypes(include = ['float64','int64']).columns
imp = SimpleImputer(missing_values=np.nan, strategy='mean')
dados_3 = imp.fit_transform(data[cat])
dados_3
```

```python
dados = np.concatenate((dados_3,dados_2), axis=1)
dados
```

```python
data=pd.DataFrame(dados,columns=['A2','A3','A8','A11','A14','A15','A1','A4','A5','A6','A7','A9','A10','A12','A13','CLASSE'])
data['A2'] = data['A2'].apply(float)
data['A3'] = data['A3'].apply(float)
data['A8'] = data['A8'].apply(float)
data['A11'] = data['A11'].apply(int)
data['A14'] = data['A14'].apply(float)
data['A15'] = data['A15'].apply(int)
data.info()
data
```

```python
cat = data.select_dtypes(include = ['object']).columns
cat_ex = []
for col in cat:
    if data[col].nunique() >= 3:
        cat_ex.append(col)
cat_ex
```

```python
data = data.drop(cat_ex, axis=1)
data
```

```python
data_categorico = data.select_dtypes(include=['object'])

from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder()

data_encoded = encoder.fit_transform(data_categorico)

colunas_encoded = encoder.get_feature_names_out(data_categorico.columns)

data_encoded_df = pd.DataFrame(data_encoded.toarray(), columns=colunas_encoded)

# Concatena os dataframes original e codificado
data_final = pd.concat([data, data_encoded_df], axis=1)

# Remove as colunas originais que foram codificadas
data_final = data_final.drop(data_categorico.columns, axis=1)

data_final
```

```python
from sklearn.neighbors import LocalOutlierFactor


lof = LocalOutlierFactor(n_neighbors=20, contamination=0.05)

anomalias = lof.fit_predict(data_final)
data_final['anomalias'] = anomalias
data_final
```

```python
lofs_f = lof.negative_outlier_factor_
lofs_f
```

```python
anomalias_df = data_final[data_final['anomalias'] == -1]
anomalias_df
```

```python
import pandas as pd

indice = anomalias_df.index.values
print('Outliers')
print(indice)
data_final = data_final.drop('anomalias', axis=1)
data_final = data_final.drop(indice)
data_final
```

```python
df = data_final.drop(columns="CLASSE_neg")
df = df.drop(columns="CLASSE_pos")
df
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
df_padronizado = scaler.fit_transform(df)
```

```python
pca = PCA(n_components=0.3)
projected = pca.fit_transform(df_padronizado)

print("Componentes Principais")
print(pca.components_)

print("\n")
print("Variâncias")
print(pca.explained_variance_ratio_)

component_names = ['component {}'.format(i) for i in range(len(pca.components_))]

print("\n")
components_df = pd.DataFrame(data=pca.components_,index=component_names,columns=df.columns)
components_df.head()
```
